```python
import matplotlib.pyplot as plt #그래프 그리는 용도로 사용되는 라이브러리
import pandas as pd #데이터를 가져오는 라이브러리
plt.rc('font', family='Malgun Gothic') #한글 폰트 사용
plt.rc('axes', unicode_minus=False)

#항성 표 엑셀 파일에서 데이터 가져오기

df = pd.DataFrame
df = pd.read_excel("항성표.xlsx") 
df

#항성 표에서 종류를 통해 데이터 나누기

DF_SD = df[(df['종류'] == 'SD')]
DF_MS = df[(df['종류'] == 'MS')]
DF_SbG = df[(df['종류'] == 'SbG')]
DF_G = df[(df['종류'] == 'G')]
DF_SpG = df[(df['종류'] == 'SpG')]
DF_WD = df[(df['종류'] == 'WD')]



fig, ax = plt.subplots()

#검은 배경 설정

plt.style.use('dark_background')

#x, y축 반전 및 설명 추가
ax.invert_xaxis()
ax.invert_yaxis()
ax.set_xlabel('온도(K)')
ax.set_ylabel('절대등급')
ax.set_title('H-R도')

#그래프에 데이터 추가
#준왜성 : 노랑
#주계열성 : 밝은 주황
#준거성 : 주황
#거성 : 어두운 빨강
#초거성 : 빨강
#백색 왜성 : 하양
ax.plot(DF_SD['온도'], DF_SD['절대등급'], marker = 'o', ls = 'none', color = 'yellow')
ax.plot(DF_MS['온도'], DF_MS['절대등급'], marker = 'o', ls = 'none', color = 'orange')
ax.plot(DF_SbG['온도'], DF_SbG['절대등급'], marker = 'o', ls = 'none', color = '#FF3300')
ax.plot(DF_G['온도'], DF_G['절대등급'], marker = 'o', ls = 'none', color = '#660000')
ax.plot(DF_SpG['온도'], DF_SpG['절대등급'], marker = 'o', ls = 'none', color = '#FF0000')
ax.plot(DF_WD['온도'], DF_WD['절대등급'], marker = 'o', ls = 'none', color = 'white')

plt.show()
```

```python
# import matplotlib.pyplot as plt #그래프 그리는 용도로 사용되는 라이브러리
import pandas as pd #데이터를 가져오는 라이브러리
plt.rc('font', family='Malgun Gothic') #한글 폰트 사용
plt.rc('axes', unicode_minus=False)

#항성 표 엑셀 파일에서 데이터 가져오기

df_Galaxy = pd.DataFrame
df_Galaxy = pd.read_excel('은하표.xlsx')
df_Galaxy

fig, ax = plt.subplots()

#검은 배경 설정

plt.style.use('dark_background')

#x, y축 설명 추가
ax.set_xlabel('거리(Mly)')
ax.set_ylabel('속도(km/s)')
ax.set_title('거리-속도 표')

#표 그리기(시선속도는 절대값으로 계산한다.)
ax.errorbar(df_Galaxy['거리'], df_Galaxy['시선속도'].abs(), df_Galaxy['시선속도오차'], marker = 'o', ls = 'none', ms = 1, color = 'white')

#평균점(빨간 x)과 원점에서 중점을 잇는 직선 표시
ax.plot(df_Galaxy['거리'].mean(), df_Galaxy['시선속도'].abs().mean(), marker = 'x', ls = 'none', ms = 4, color = 'red')
ax.plot([0,df_Galaxy['거리'].mean()*2], [0,df_Galaxy['시선속도'].abs().mean()*2], ls = '--', ms = 1, color = 'red', alpha = 0.5)
#허블 상수 구하기

#광년으로 계산한 허블 상수(km/s/Mly)
Hly = df_Galaxy['시선속도'].abs().mean()/df_Galaxy['거리'].mean()

#파섹으로 계산한 허블 상수(km/s/Mpc) (이때 1ly = 0.3066013938pc로 가정한다.)
Hpc = df_Galaxy['시선속도'].abs().mean()/df_Galaxy['거리'].mean()/0.3066013938
plt.show()
print('광년으로 계산한 허블상수')
print(f'{Hly}km/s/Mly')
print('파섹으로 계산한 허블상수(1ly = 0.3066013938pc)')
print(f'{Hpc}km/s/Mpc')
```
